## TASK A: Create a pd.Series

Create a pd.Series from a Python list of your favorite 5 numbers.
- Print the index and values separately.


```python
import pandas as pd
import numpy as np

# 1. Create a Series
s = pd.Series([10, 20, 30, 40, 50])
print("Index:", s.index)
print("Values:", s.values)
```

## TASK B: create a DataFrame

Create a DataFrame with 3 columns: "Name", "Age", "University". Add at least 5 rows.

Print:
- the first two rows,
- the "Age" column as a Series,
- the "Name" and "University" columns as a DataFrame.
- Select the last row using .iloc and print it.

```python
data = {
    "Name": ["Alice", "Bob", "Charlie", "Diana", "Ethan"],
    "Age": [24, 30, 22, 28, 35],
    "University": ["Unipi", "IMT", "Unicam", "Luiss", "Unibo"]
}
df = pd.DataFrame(data)

# 2. Retrieve rows & columns
print(df.head(2))             # first two rows
print("---------------------------")
print(df["Age"])              # Age column (Series)
print("---------------------------")
print(df[["Name", "University"]])   # Name and City columns (DataFrame)
print("---------------------------")

# 3. Select last row using iloc
print(df.iloc[-1])
```

## TASK C: play with missing and Duplicate Data
- Add a new row with missing values (NaN).
- Use .isna() to check for missing data.
- Replace missing "Age" with the mean age.
- Duplicate one row intentionally.
- Use .duplicated() to detect it, and .drop_duplicates() to remove it.

```python
# 1. Add a row with NaN
df.loc[len(df)] = ["Frank", np.nan, 50]  # "Name", "Age", "Age_in_10_years"
print(df)

# Detect NaN
print(df.isna())

print("---------------------------")
# Fill NaN in Age with mean
df["Age"].fillna(df["Age"].mean(), inplace=True)

# 2. Add duplicate row
df.loc[len(df)] = ["Alice", 24, 34]
print(df)
print("---------------------------")

# Detect duplicates
print(df.duplicated())
print("---------------------------")

# Remove duplicates
df = df.drop_duplicates()
print(df)
```
