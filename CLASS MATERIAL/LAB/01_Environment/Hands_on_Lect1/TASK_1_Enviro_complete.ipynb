### <font color=blue>Step 1: install Anaconda </font>
Download Anaconda from the [official website](https://www.anaconda.com/products/individual) and follow the [instructions](https://docs.anaconda.com/anaconda/install/) for your OS.

**conda** is the package and environment manager provided with Anaconda. From the [conda website](https://conda.io/en/latest/):
> Conda is an open source **package management system** and **environment management system** that runs on Windows, macOS and Linux. Conda quickly installs, runs and updates packages and their dependencies. Conda easily creates, saves, loads and switches between environments on your local computer. It was created for Python programs, but it can package and distribute software for any language.



### <font color=blue>Step 2: Create a conda environment</font>


Open a Unix shell or Anaconda prompt

- Enable conda-forge channel for conda package manager
```bash
$ conda config --append channels conda-forge
```
- Create the environment named "DMML_AY2526" using the following command. It will also install (hopefully all) the required packages. You will be asked to confirm the installation.
```bash
$ conda create --name DMML_AY2526 python=3.10 jupyter pandas openpyxl seaborn matplotlib scikit-learn plotly
$ conda activate DMML_AY2526 
```



### <font color=blue>Step 3: Check everything is ok</font>


Open a Unix shell or Anaconda prompt

1. Get the list of available environments (there should also be DMML_AY2526)
```bash
$ conda env list
```
2. Activate your newly created environment
```bash
$ conda activate DMML_AY2526
```

3. Start the jupyter notebook server
```
(DMML_AY2526)$ jupyter notebook
```

You should get something like this, in your browser:

![jupyter-home.png](https://www.dataquest.io/wp-content/uploads/2019/01/jupyter-dashboard.jpg)

4. If you need to go back to the "system context":

```bash
(DMML_AY2526)$ conda deactivate
```
