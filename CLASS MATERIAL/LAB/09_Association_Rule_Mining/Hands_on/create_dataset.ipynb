```python
import random

items = ["milk", "bread", "diapers", "beer", "coffee", "water"]

def generate_transaction():
    transaction = []
    
    # Base probabilities
    probs = {
        "milk": 0.4,
        "bread": 0.5,
        "diapers": 0.2,
        "beer": 0.25,
        "coffee": 0.3,
        "water": 0.35
    }
    
    for item, p in probs.items():
        if random.random() < p:
            transaction.append(item)
    
    # Inject the secret rule
    if "diapers" in transaction and "beer" not in transaction:
        if random.random() < 0.75:
            transaction.append("beer")
    
    return transaction

transactions = [generate_transaction() for _ in range(300)]
```

```python
# perform one hot encoding
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd

te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df = pd.DataFrame(te_ary, columns=te.columns_)
df.head()


df.to_csv("market_dataset.csv")
```
