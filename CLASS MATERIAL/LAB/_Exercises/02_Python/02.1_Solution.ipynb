# Computations

If you run a 10 kilometer race in 42 minutes 42 seconds,
- What is your average pace (time per mile in minutes and seconds)? 
- What is your average speed in miles per hour?

*Hint*: there are 1.61 kilometers in a mile.

````python
# convert kilometers to miles
KILOMETERS_PER_MILES = 1.61
MILES_PER_KILOMETERS = 1 / KILOMETERS_PER_MILES
space_km = 10

space_miles = space_km * MILES_PER_KILOMETERS
space_miles

````

````python
# convert time to seconds
minutes = 42
seconds = 42
total_seconds = minutes * 60 + seconds
total_seconds
````

````python
# what is your average time (seconds) per mile
avg_time_per_mile = total_seconds / space_miles
avg_time_per_mile
````

````python
# what is your average time (minutes) per mile
avg_minutes_per_mile = avg_time_per_mile // 60
avg_seconds_per_mile = avg_time_per_mile % 60
print(f'{avg_minutes_per_mile} minutes and {avg_seconds_per_mile} seconds')
````

````python
# miles per hour
miles_per_hour = 60 * 60 / avg_time_per_mile
miles_per_hour
````

# Lists
Consider the following list:
```python
planets = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", "Neptune", "Uranus"]
```
- restore Pluto's status as a planet and add it to the list
- use the slice operator to pull out the rocky planets (Mercury, Venus, Earth, Mars) as a new list 
- write a program to identify the first rocky planet in alphabetical order





````python
planets = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", "Neptune", "Uranus"]
planets
````

````python
planets.append('Pluto')
planets
````

````python
mvem = planets[:4] # equivalent to planets[0:4]
mvem
````

````python
min(mvem)
````

````python
print(mvem) # original list
mvem.sort()
print(mvem[0])
print(mvem) # original list altered in place
````

````python
mvem = planets[:4] # restore the slicing
print(mvem) # original list
print(sorted(mvem)[0])
print(mvem) # original list not altered
````
