- write a program that, given a string `'abcdefghi'`, prints the following output 
```python
Char 1: a
Char 2: b
Char 3: c
...    
```

````python
string = 'abcdefghi'
for enu,char in enumerate(string):
    print(f'Char {enu+1}: {char}')
````

````python
# different string formatting
for enu,char in enumerate(string):
    print('Char %d: %s'%(enu+1,char))
````

````python
# different string formatting
for enu,char in enumerate(string):
    print('Char {index}: {character}'.format(index=enu+1, character=char))
````

````python
string = input('Insert one string: ')
for enu, char in enumerate(string):
    print(f'Char {enu+1}: {char}')
````

# Loops

- given the following list of numbers, create a dictionary hosting the number of occurrences of each distinct item in the list
```python
numbers = [1, 2, 2, 7, 4, 9, 7, 4, 8, 2, 4, 4, 6, 6, 7, 2, 4, 1, 2, 9, 8, 1, 8, 7, 9]
```



````python
numbers = [1, 2, 2, 7, 4, 9, 7, 4, 8, 2, 4, 4, 6, 6, 7, 2, 4, 1, 2, 9, 8, 1, 8, 7, 9]
counts = {}
for n in numbers:
    counts[n] = counts.get(n, 0) + 1
print(counts)
````

````python
# alternative, with exception
numbers = [1, 2, 2, 7, 4, 9, 7, 4, 8, 2, 4, 4, 6, 6, 7, 2, 4, 1, 2, 9, 8, 1, 8, 7, 9]
counts = {}
for x in numbers:
    try:
        counts[x] += 1
    except KeyError:
        counts[x] = 1
print(counts)
````

````python
# alternative
counts = {}
for x in set(numbers):
    counts[x] = numbers.count(x)
counts
````

# Loops and conditions

- write a program that, given a string provided by the user (use `input()` method), prints the length of the string. However, the process must terminate if the user inputs the string `'exit'`

````python
string = ''
while string != 'exit':
    string = input('Insert a string: ')
    print(len(string))
````

````python
while True:
    string = input('Insert a string: ')
    if string == 'exit':
        break
    print(len(string))
````

# Loops and conditions
- write a program that, given an integer $n$ provided by the user (use `input()` method), prints all the prime numbers $i$, s.t. $i\leq n$ 

````python
num = input('Insert a number: ')
n = 2
while n<int(num):
    is_prime = True
    for k in range(2,n):
        if n%k==0:
            is_prime = False
    if is_prime:
        print(n, end = ', ')
    n += 1 
````
