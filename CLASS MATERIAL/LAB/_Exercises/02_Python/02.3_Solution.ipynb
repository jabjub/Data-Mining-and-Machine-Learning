# Functions
- define a function `inflate()` to inflate the prices of goods by a given default increase multiplier.
- the goods prices are passed as a dictionary
- the increase multiplier is passed as a float, and should have a default increase of 5%


````python
def inflate(prices, multiplier = 1.05):
    """ inflate prices: the dictionary is modified in place
    
    Args:
        prices: dictionary
        multiplier: inflate rate as a multiplier (default = 5%)
    """
    for x in prices:
        prices[x] *= multiplier 
        prices[x] = round(prices[x],2) # <- not critical but looks nicer
        
goods = {'car': 1000.00,
         'paper': 50.50}
goods
````

````python
inflate(goods) 
goods
````

````python
inflate(goods, 2)
goods
````

````python
def inflate2(prices, multiplier = 1.05):
    """ inflate prices: the original dictionary is not modified
    
    Args:
        prices: dictionary
        multiplier: inflate rate as a multiplier (default = 5%)
    """
    return {k:round(v*multiplier,2) for k,v in prices.items()}

goods = {'car': 1000.00,
         'paper': 50.50}
````

````python
inflate2(goods)
````

````python
inflate2(goods, 2)
````

````python
goods
````

# Recursive functions
Python supports a construct for which a defined function can call itself.
- make use of a recursive function to compute the factorial of a number
- make use of a recursive function to compute the sum of the elements in a list

````python
def factorial(n):
    if n == 1 or n == 0:
        return 1
    return n*factorial(n-1)
````

````python
factorial(4)
````

````python
def sum_seq(seq):
    item = seq.pop() # return the last element and modify the list inplace
    if len(seq)==0:
        return item 
    return item+sum_seq(seq)
````

````python
sum_seq([1,2,3,44])
````

````python
# alternative, good also for tuples
def sum_seq_2(seq):
    item = seq[0] 
    if len(seq[1:])==0:
        return item 
    return item+sum_seq_2(seq[1:])
````

````python
sum_seq_2([1,2,3,44])
````

````python
sum_seq_2((1,2,3,44))
````

# Doodle: A function to schedule a meeting


> In a few sentences, describe an algorithm for scheduling a meeting with $N$ people. Assume that the meeting must take place on a given day, and that you have a list of available times (hour-long blocks) from each person. Your algorithm should report a time that works for everyone, or report that no such time exists.


Here is a sample solution:

> Count the number of occurrences of each candidate meeting time across the $N$ people's lists. If at least one candidate meeting time occurred $N$ times, pick the earliest as the meeting time. Otherwise, inform the group that no time worked for everyone.

* We will represent the individuals' schedules as a dictionary of lists: the dictionary key is the individual's name, and the list contains their available meeting times (represented as strings).

* Implement this solution in Python code as a function `find_slot( )` that takes the set of schedules as an argument and returns the earliest meeting time that work for everyone. If there is no meeting time working for everyone, inform the group that no time worked for everyone.


````python
# execute this cell to define
schedules_1 = {
    "Alex": ["09am", "11am", "01pm", "05pm"],
    "Beth": ["11am", "12pm", "01pm"],
    "Carl": ["11am", '01pm'],
}
````

````python
# execute this cell to define
schedules_2 = {
    "Alex": ["09am", "11pm", "01pm", "05pm"],
    "Beth": ["11pm", "12pm", "01pm"],
    "Carl": ["11pm", '01pm'],
}
````

````python
# execute this cell to define
schedules_3 = {
    "Alex": ["09am", "11am", "01pm", "05pm"],
    "Beth": ["10am", "12pm", "01pm"],
    "Carl": ["11am", '02pm'],
}
````

````python
# execute this cell to define
schedules_4 = {
    "Alex": ["09am", "11pm", "01pm", "05pm"],
    "Beth": ["11pm", "12pm", "01pm"],
    "Carl": ["11pm", '00pm'],
}
````

````python
def find_slot(schedules):
    meeting_times = []
    for person,availability in schedules.items():
        meeting_times.extend(availability)
    candidate = []
    for unique in set(meeting_times):
        count_times = meeting_times.count(unique)
        if count_times == len(schedules):
            converted_unique = int(unique[:2]) + (12 if unique[2:] == 'pm' else 0)
            candidate.append(converted_unique)
    if len(candidate) == 0:
        choice = "Sorry, can't make it work!"
    else:
        choice = sorted(candidate)[0]   
    # print('candidate', candidate) #debug
    return choice
````
