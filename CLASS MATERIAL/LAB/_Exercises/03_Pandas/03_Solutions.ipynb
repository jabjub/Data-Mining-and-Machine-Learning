```python
counter = 0
for enu,x in df.groupby('item_name')[['dollars','quantity']]:
    print(enu)
    if counter < 3:
        display(x)
    x['unit_cost'] = x['dollars']/x['quantity']
    print(len(x['unit_cost'].unique()))
    counter += 1
```

```python
df = pd.read_csv(os.path.join('dataset','healthCovid19.csv'),index_col = 0)
df.head(10)
```

Note that it will raise an error if the country name is not found in the library look-up table


```python
import pycountry_convert as pc

def to_code(country):
    try:
        country_code = pc.country_name_to_country_alpha2(country, cn_name_format="default")
    except KeyError:
        country_code = None
    return country_code

def to_continent(code):
    try:
        continent_name = pc.country_alpha2_to_continent_code(code)
    except KeyError:
        continent_name = None
    return continent_name
```

```python
df['Country_Code'] = df.index.to_series().apply(lambda x: to_code(x))
df['Continent'] = df['Country_Code'].apply(lambda x: to_continent(x))
df
```

```python
df.isna().sum(axis = 0)
```

The code column can be safely deleted

```python
df = df.drop(columns = 'Country_Code',axis = 1)
df
```

Now we can look at aggregated information, grouping countries by continent.

```python
df.info()
```

```python
aggregated_df = df.groupby('Continent').mean(numeric_only = True)
```

```python
aggregated_df['Population'].plot(kind = 'bar')
plt.title('Population')
```

Which continents have the highest ratio reported for Covid19 deaths?

```python
aggregated_df['Deaths'].plot(kind = 'bar')
plt.title('Death percentage')
```

Which is the state with the highest rate of death?

```python
df.loc[df['Deaths'].idxmax()]
```

```python
df.sort_values(by='Deaths', ascending=False)
```

```python
df.sort_values(by='Deaths', ascending=False).head(8).T
```

Compare the distribution of Deaths rates among European and North American countries.

```python
deathNA = df.groupby('Continent').get_group('NA').Deaths
deathEU = df.groupby('Continent').get_group('EU').Deaths
```

```python
plt.hist(deathNA, 10, alpha=0.5, label='NA')
plt.hist(deathEU, 10, alpha=0.5, label='EU')
plt.legend(loc='upper right')
plt.xlabel('Death percentage')
plt.ylabel('# countries')
plt.show()
```
