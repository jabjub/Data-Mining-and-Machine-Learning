## Perform the required data preprocessing steps 
- handle missing values
- obtain a numeric representation of the categorical variables.


````python
df = df.drop('id',axis = 1)
````

### Step 0: split X and y

````python
X_train = df.drop("satisfaction", axis = 1)
y_train = df['satisfaction']
````

### Step 1: handling Class column

````python
diz_class = {'Eco': 0, 'Eco Plus': 1, 'Business': 2}
````

````python
X_train['Class'] = X_train['Class'].apply(lambda x: diz_class[x])
ordinal_variables.append('Class')
nominal_variables.remove('Class')
````

### Step 2: encode boolean variables

````python
diz_loyalty = {'Loyal Customer': 1,
               'disloyal Customer': 0}
diz_gender = {'Male': 1,
              'Female': 0}
diz_type = {'Personal Travel': 0,
            'Business travel': 1}

X_train['Customer Type'] = X_train['Customer Type'].apply(lambda x: diz_loyalty[x])
X_train['Gender'] = X_train['Gender'].apply(lambda x: diz_gender[x])
X_train['Type of Travel'] = X_train['Type of Travel'].apply(lambda x: diz_type[x])
````

### Step 3: handling missing values in arrival delay

There are several options for handling missing values:
1. dropping the instances (the percentage is quite low) (i will need a policy at inference time, anyway).
2. replacing missing values with the median value (mean would be affected by the presence of outliers). 
3. filling the missing "Arrival Delay" based on the actual "Departure Delay", since we have verified that there exists a linear relationships

It is also ok to drop the whole column (Arrival Delay in Minutes), as it is highly correlated with Departure delay. 

````python
X_train = X_train.drop('Arrival Delay in Minutes', axis = 1)
numerical_variables.remove('Arrival Delay in Minutes')
````

### Step 4: handling missing values in surveys

I opt for replacing missing values with the mode (most frequent value).

````python
from sklearn.impute import SimpleImputer
sim = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
sim.fit(X_train)
X_train_imputed = pd.DataFrame(sim.transform(X_train))
X_train_imputed.columns=X_train.columns
X_train_imputed
````
